```python
class Aluno: #Eles servem pra armazenar os dados dentro do objeto o INIT é o método construtor. Os dois tracinhos (--)indicam que são privados. 
    def __init__(self, nome, matricula, curso): # a variável notas começacomo uma lista vazia, onde vai ser guardada as notas do aluno.
        self.__nome = nome 
        self.__matricula = matricula 
        self.__curso = curso 
        self.__notas = [] #

    # Getter para o nome:
    def get_nome(self):#Serve para retornar né, voltar pro nome do aluno com segurança,
        return self.__nome# o que permite acessar o atributo privado

    # Getter para a matrícula
    def get_matricula(self):# Serve pra retornar a matricula do aluno.
        return self.__matricula

    def adicionar_nota(self, nota): # Adiciona a nota a lista de notas vazia la de cima, verifica se
        if 0 <= nota <= 10:         # a nota ta entre 0 e 10, ai se tiver correta a nota é adicionada usando o append(nota)pra adicionar
            self.__notas.append(nota)

    def calcular_media(self): #Serve para calcular a média das notas, sum soma todas as notas e len conta quantas notas tem, ai depois divide e soma.
        return sum(self.__notas) / len(self.__notas) if self.__notas else 0  #e evita erro caso a lista esteja vazia, retornando 0.

    def mostrar_dados(self): #serve pra mostrar as informações dos alunos
        print(f"Nome: {self.__nome} | Matrícula: {self.__matricula}")
        print(f"Média: {self.calcular_media():.2f}")
        self.__curso.descricao()
```

```python
# cadastro.py

from curso import Curso #Serve p importar as classes do curso  aluno de outros arquivos 
from aluno import Aluno


def cadastrar_cursos(): #Utilizado para cadastrar cursos. Organiza o processo da criação e dos cursos digitados pelos usuarios.
    cursos = [] #lista vazia pra guardadr todos os cursos.
    quantidade = int(input("Quantos cursos deseja cadastrar? "))

    for i in range(quantidade):# faz repetir o cadastro conforme a quantidade que o usuário botou la.
        print(f"\n📘 Cadastro do curso {i+1}") #vão pedir o nome e a duração(número inteiro)
        nome = input("Nome do curso: ")
        duracao = int(input("Duração (em semestres): "))
        cursos.append(Curso(nome, duracao)) #cria um objeto da classe curso e anexa usando o append

    return cursos #retoma a lista com todos os cursos cadastrados.


def cadastrar_alunos(cursos): #serve pra cadastrar alunos e colocar eles em um curso que ja existe
    alunos = [] # cria uma lista vazia onde armazena os alubnos cadastrados
    quantidade = int(input("\nQuantos alunos deseja cadastrar? ")) #perguntar quantos alunos o usuário deseja cadastrar

    for i in range(quantidade): #repete o processso de cadastro pra cada aluno
        print(f"\n📌 Cadastro do aluno {i+1}") #pedem o nome e a matricula do aluno
        nome = input("Nome: ")
        matricula = input("Matrícula (8 a 10 dígitos): ")

        while True: #cria um lop infinito que vai sndo execuitado até que a pessoa escolha o curso que sej válido
            escolha = input("Digite o número do curso: ") #tem que ser válido
            if escolha.isdigit() and 1 <= int(escolha) <= len(cursos): #verificar se o valor digitado é um número e se ta dentro da quantidade de curso
                curso_escolhido = cursos[int(escolha) - 1] #seleciona o curso correspondente
                break #encerra o laço while quand a escolha é válida
            else:
                print("⚠️ Escolha inválida.") #mostra uma mensagem de erro caso seja inválida

        aluno = Aluno(nome, matricula, curso_escolhido) #cria objet da classe aluno e adicionam a lista de números de alunos cadastrados
        alunos.append(aluno)

    return alunos#retorna a lista de alunos cadastrados


def adicionar_notas(alunos): #adiciona notas aos alunos cadastrados
    for aluno in alunos:#faz o programa percorrese todos os alunos da lista
        print(f"\nAdicionando notas para {aluno.get_nome()}:")#armazena o nome do curso
        for i in range(2): #cada aluno vai receber duas notas
            # O programa entra em um estado de "espera" e só sairá dele (através do comando break)
            while True:
                entrada = input( #solicita ao usuário a nota do aluno
                    f"Digite a nota {i+1} para {aluno.get_nome()}: ")
                if entrada.strip() == "":#verifica se o usuário deixou o campo vazio e o strip vai remover espaços antes da verificação
                    print("⚠️ Entrada vazia. Digite uma nota válida.") #exibido quando nenhuma nota é digitada
                    continue #aqui o programa volta la no inicio do laçopra pedir a norta de novo
                try: #tenta converter a entrada em um número decimal para permitir notas com pontos e virgulas
                    nota = float(entrada)
                    if 0 <= nota <= 10: #verifica se a nota ta entre 10 e 0
                        aluno.adicionar_nota(nota) #adiciona a nota ao aluno usando o métodocriado la em cima na classe aluno
                        break #de novoencerra o laço quando se tem uma nota válida
                    else:
                        print("⚠️ A nota deve estar entre 0 e 10.")#aparece quando se tem uma nota fora do limite que foi permitido
                except ValueError:#captura erros caso o usuário digite algo que não seja um número
                    print("⚠️ Valor inválido. Digite um número.") #mostra uma mensagem dizendo que esta inválido
```

```python
# curso.py

class Curso:#cria uma classe chamada curso
    def __init__(self, nome, duracao):#aqui tem o método construtor de novo, que é responsável por inicializarod dados do curso quando se cria umobjeto
        self.nome = nome #armazena o nome do curso
        self.duracao = duracao #armazena a duraçãodo curso

    def descricao(self): #cria um método para mostrar as informações do curso
        print(f"Curso: {self.nome} - Duração: {self.duracao} semestres")#exibe na tela o nome do curso e a duração em semestres
```
